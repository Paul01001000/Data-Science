```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/paul8/Downloads/clustering_exercise.csv")
data.head()
```

```python
model = KMeans(n_clusters=2, random_state=42)
features = data[['meat_spending','vegetable_spending']]
model.fit(features)
pred = model.predict(features)
```

```python
plt.scatter(data.meat_spending, data.vegetable_spending,c=pred)
```

```python
confusion_matrix = pd.crosstab(data['class'], pred, rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
confusion_matrix = pd.crosstab(data['prime_member'], pred, rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```
