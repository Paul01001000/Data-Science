```python
import pandas as pd
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("C:/Users/paul8/Downloads/Fish.csv")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[["Length1","Length2","Length3","Height","Width"]],df.Weight, test_size=0.3, random_state=42)
```

```python
model = XGBRegressor()
model.fit(X_train,Y_train)
```

```python
pred_test = model.predict(X_test)
pred_train = model.predict(X_train)
```

```python
plt.scatter(Y_test,pred_test)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.title("Test Set Error")
```

```python
plt.scatter(Y_train,pred_train)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.title("Test Set Error")
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

MAE_model = mean_absolute_error(Y_test, pred_test)
MAE_baseline = mean_absolute_error(Y_test, np.repeat(Y_train.mean(), len(Y_test)))
print(MAE_model)
print(MAE_baseline)
```
