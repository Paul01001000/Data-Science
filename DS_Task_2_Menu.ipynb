```python
data = data[["Hamburger","Cheeseburger","Veggie Burger","Fries","Salad", "Nachos","Cupcake","Cookie","Soft Drink","Water","Juice","Coffee","Tea","Beer","Wine","Cocktail","Pretzel"]]
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".1f", linewidths=.2)
plt.show()
```

```python
from sklearn.cluster import KMeans

four_Means = KMeans(n_clusters=4, random_state=42)
four_Means.fit(data)

centers = four_Means.cluster_centers_

centers = pd.DataFrame(centers, columns=data.columns)
centers
```

```python
from sklearn.metrics import davies_bouldin_score
res = four_Means.predict(data)
davies_bouldin_score(data,res)
```

```python
data["menu"] = res
data.head()
```

```python
menu_zero = data[data["menu"] == 0]
menu_one = data[data["menu"] == 1]
menu_two = data[data["menu"] == 2]
menu_three = data[data["menu"] == 3]
```
